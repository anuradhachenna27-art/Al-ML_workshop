```python
1. fifo
# =========================
from queue import Queue
import threading
import time

scan_queue = Queue()
scan_queue.put("User_101")
scan_queue.put("User_102")

# =========================
# 2. DDL – Database Structure (Conceptual)
# =========================
"""
CREATE TABLE fingerprint_users (
    user_id INT PRIMARY KEY,
    name VARCHAR(50),
    fingerprint_hash VARCHAR(255)
);
"""

# =========================
# 3. DML – Existing Fingerprint Data (Simulation)
# =========================
fingerprint_database = {
    "fp_hash_abc": "User_101",
    "fp_hash_xyz": "User_102"
}

# =========================
# 4. Multithreading – Scan & Match in Parallel
# =========================
def scan_fingerprint():
    time.sleep(2)
    return "fp_hash_abc"  # simulated scan result

def match_fingerprint(fp_data):
    time.sleep(1)
    return fp_data in fingerprint_database

# =========================
# 5. Error Handling – Sensor & Matching Errors
# =========================
def fingerprint_process():
    try:
        if scan_queue.empty():
            raise Exception("No users in queue")

        user = scan_queue.get()  # FIFO applied
        print(f"Scanning fingerprint for {user}")

        # Multithreading starts
        scan_thread = threading.Thread(target=scan_fingerprint)
        scan_thread.start()
        scan_thread.join()

        fp_data = scan_fingerprint()
        if fp_data is None:
            raise ValueError("Fingerprint not detected")

        match_result = match_fingerprint(fp_data)

        # =========================
        # Output Result (YES / NO)
        # =========================
        if match_result:
            print("\033[92mYES – Access Granted\033[0m")
        else:
            print("\033[91mNO – Access Denied\033[0m")

        # =========================
        # 6. LIFO – Recent Scan Logs
        # =========================
        scan_logs.append(user)

    except Exception as e:
        print("Error:", e)

# =========================
# LIFO Stack Initialization
# =========================
scan_logs = []

# =========================
# Execute Fingerprint System
# =========================
fingerprint_process()

# =========================
# Review Last Scan (LIFO)
# =========================
if scan_logs:
    print("Last scanned user:", scan_logs.pop())
```
